# Example 1: Standard binned analysis
### Import the BinnedData class from cosipy

```python
from cosipy import BinnedData
```

### Define the BinnedData object
### Input parameters are passed from inputs.yaml

```python
analysis = BinnedData("inputs.yaml")
```

```python
analysis.read_tra()
```

### Bin the data

```python
analysis.get_binned_data()
```

### Plot the raw spectrum and lightcurve

```python
%matplotlib inline
analysis.get_raw_spectrum()
```

```python
%matplotlib inline
analysis.get_raw_lightcurve()
```

# Example 2: Some available options for the standard binned anlaysis
### In the previous step the unbinned data is saved to an hdf5 file with the read_tra() method.
### Here we will load the unbinnned data from file instead of running read_tra() again.
### We will also make binning plots.

```python
analysis = BinnedData("inputs.yaml")
analysis.get_binned_data(unbinned_data="unbinned_data.hdf5", make_binning_plots=True)
```

### The get_binned_data method writes the binned data to an hdf5 file.
### Here we will load the binnned data from file to make the raw spectrum and lightcurve
### We will also save the spectrum and lightcurve to .dat files.

```python
analysis = BinnedData("inputs.yaml")
analysis.get_raw_spectrum(binned_data="binned_data.hdf5")
analysis.get_raw_lightcurve(binned_data="binned_data.hdf5")
```
